```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv("submission_07_seed1.csv")
#file2 = pd.read_csv("submission_08_seed2.csv")
#file3 = pd.read_csv("submission_08_seed3.csv")
#file4 = pd.read_csv("submission_08_seed4.csv")
#file5 = pd.read_csv("submission_08_seed5.csv")
#file6 = pd.read_csv("submission_08_seed6.csv")
#file7 = pd.read_csv("submission_08_seed7.csv")
#file8 = pd.read_csv("submission_08_seed8.csv")
#file9 = pd.read_csv("submission_08_seed9.csv")
#file10 = pd.read_csv("submission_08_seed10.csv")
```

```python
utsunomiya = file1["pollen_utsunomiya"].copy()

#4の倍数にそろえる
for i in range(1000):
    utsunomiya[(utsunomiya > i*4-0) & (utsunomiya <= i*4+4)] = i*4 #小さく丸めたほうが精度高いようなので過小評価する
utsunomiya[utsunomiya < 0] = 0
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_utsunomiya"], label="file1")
plt.plot(utsunomiya, label="sub")

plt.legend()
```

```python
chiba = file1["pollen_chiba"].copy()

#4の倍数にそろえる
for i in range(1000):
    chiba[(chiba > i*4-2) & (chiba <= i*4+2)] = i*4
chiba[chiba < 0] = 0
```

```python
plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_chiba"], label="file1")
plt.plot(chiba, label="sub")

plt.legend()
```

```python
tokyo = file1["pollen_tokyo"].copy()

#4の倍数にそろえる
for i in range(1000):
    tokyo[(tokyo > i*4-0) & (tokyo <= i*4+4)] = i*4 #小さく丸めたほうが精度高いようなので過小評価する
tokyo[tokyo < 0] = 0
```

```python
plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_tokyo"], label="file1")
plt.plot(tokyo, label="sub")

plt.legend()
```

```python
df_submission = pd.read_csv("sample_submission.csv")
```

```python
df_submission["pollen_utsunomiya"] = utsunomiya
df_submission["pollen_chiba"] = chiba
df_submission["pollen_tokyo"] = tokyo
df_submission
```
