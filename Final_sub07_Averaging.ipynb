```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv("submission_07_seed1.csv")
file2 = pd.read_csv("submission_07_seed2.csv")
file3 = pd.read_csv("submission_07_seed3.csv")
file4 = pd.read_csv("submission_07_seed4.csv")
file5 = pd.read_csv("submission_07_seed5.csv")
file6 = pd.read_csv("submission_07_seed6.csv")
file7 = pd.read_csv("submission_07_seed7.csv")
file8 = pd.read_csv("submission_07_seed8.csv")
file9 = pd.read_csv("submission_07_seed9.csv")
file10 = pd.read_csv("submission_07_seed10.csv")
```

```python
utsunomiya = (file1["pollen_utsunomiya"]+file2["pollen_utsunomiya"]+file3["pollen_utsunomiya"]+file4["pollen_utsunomiya"]+file5["pollen_utsunomiya"]
              +file6["pollen_utsunomiya"]+file7["pollen_utsunomiya"]+file8["pollen_utsunomiya"]+file9["pollen_utsunomiya"]+file10["pollen_utsunomiya"])/10

#4の倍数にそろえる
for i in range(1000):
    utsunomiya[(utsunomiya > i*4-2) & (utsunomiya <= i*4+2)] = i*4
utsunomiya[utsunomiya < 0] = 0
```

```python
#### 後で削除 ##########

import matplotlib.pyplot as plt


plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_utsunomiya"], label="file1")
plt.plot(file2["pollen_utsunomiya"], label="file2")
plt.plot(file3["pollen_utsunomiya"], label="file3")
plt.plot(file4["pollen_utsunomiya"], label="file4")
plt.plot(file5["pollen_utsunomiya"], label="file5")
plt.plot(file6["pollen_utsunomiya"], label="file6")
plt.plot(file7["pollen_utsunomiya"], label="file7")
plt.plot(file8["pollen_utsunomiya"], label="file8")
plt.plot(file9["pollen_utsunomiya"], label="file9")
plt.plot(file10["pollen_utsunomiya"], label="file10")
plt.plot(utsunomiya, label="sub")

plt.legend()
```

```python
chiba = (file1["pollen_chiba"]+file2["pollen_chiba"]+file3["pollen_chiba"]+file4["pollen_chiba"]+file5["pollen_chiba"]
         +file6["pollen_chiba"]+file7["pollen_chiba"]+file8["pollen_chiba"]+file9["pollen_chiba"]+file10["pollen_chiba"])/10


#4の倍数にそろえる
for i in range(1000):
    chiba[(chiba > i*4-2) & (chiba <= i*4+2)] = i*4
chiba[chiba < 0] = 0
```

```python
#### 後で削除 ##########

plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_chiba"], label="file1")
plt.plot(file2["pollen_chiba"], label="file2")
plt.plot(file3["pollen_chiba"], label="file3")
plt.plot(file4["pollen_chiba"], label="file4")
plt.plot(file5["pollen_chiba"], label="file5")
plt.plot(file6["pollen_chiba"], label="file6")
plt.plot(file7["pollen_chiba"], label="file7")
plt.plot(file8["pollen_chiba"], label="file8")
plt.plot(file9["pollen_chiba"], label="file9")
plt.plot(file10["pollen_chiba"], label="file10")
plt.plot(chiba, label="sub")

plt.legend()
```

```python
tokyo = (file1["pollen_tokyo"]+file2["pollen_tokyo"]+file3["pollen_tokyo"]+file4["pollen_tokyo"]+file5["pollen_tokyo"]
         +file6["pollen_tokyo"]+file7["pollen_tokyo"]+file8["pollen_tokyo"]+file9["pollen_tokyo"]+file10["pollen_tokyo"])/10


#4の倍数にそろえる
for i in range(1000):
    tokyo[(tokyo > i*4-0) & (tokyo <= i*4+4)] = i*4 #小さく丸めたほうが精度高いようなので過小評価する
tokyo[tokyo < 0] = 0
```

```python
#### 後で削除 ##########

plt.figure(figsize=(12, 3))
plt.plot(file1["pollen_tokyo"], label="file1")
plt.plot(file2["pollen_tokyo"], label="file2")
plt.plot(file3["pollen_tokyo"], label="file3")
plt.plot(file4["pollen_tokyo"], label="file4")
plt.plot(file5["pollen_tokyo"], label="file5")
plt.plot(file6["pollen_tokyo"], label="file6")
plt.plot(file7["pollen_tokyo"], label="file7")
plt.plot(file8["pollen_tokyo"], label="file8")
plt.plot(file9["pollen_tokyo"], label="file9")
plt.plot(file10["pollen_tokyo"], label="file10")
plt.plot(tokyo, label="sub")

plt.legend()
```

```python
df_submission = pd.read_csv("sample_submission.csv")
```

```python
df_submission["pollen_utsunomiya"] = utsunomiya
df_submission["pollen_chiba"] = chiba
df_submission["pollen_tokyo"] = tokyo
df_submission
```
